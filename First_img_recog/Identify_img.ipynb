```python
import numpy as np
import matplotlib.image as mpimg    # for reading in images
import matplotlib.pyplot as plt
import cv2     # computer img library
```

```python
# Read in the image
image = mpimg.imread("identify_img.jpg")

# Print out the dimension of the image
# Height, width, color channels respectively
print(f"Image dimension {image.shape}")
```

```python
# Strip colors to grayscale
gray_img = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
plt.imshow(gray_img, cmap="gray")
```

```python
# print specific grayscale pixel vales

x = 140
y = 100
pixel_pos = gray_img[y, x]
print(pixel_pos)
```
