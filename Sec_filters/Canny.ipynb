```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
#Load the image
image = cv2.imread("shirou.jpg")

#Convert image to greyscale
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.imshow(image)
```

```python
#Convert to greyscale
image_gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)

plt.imshow(image_gray, cmap="gray")
```

```python
low_canny = cv2.Canny(image_gray, 30, 120)
high_canny = cv2.Canny(image_gray, 120, 240)

f, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 10))

ax1.set_title("Captures more frequencies")
ax1.imshow(low_canny, cmap="gray")

ax2.set_title("Captures tight frequencies")
ax2.imshow(high_canny, cmap="gray")
```
